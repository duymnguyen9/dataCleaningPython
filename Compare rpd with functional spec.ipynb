```python
import pandas as pd
```

```python
anyClassDF = pd.read_csv('doc/AnyClassRPD.csv',header=None, names=['Folder', 'Column']);
anyClassFS = pd.read_csv('doc/AnyClassFuncSep.csv',header=None, names=['Folder', 'Column'])
#Remove Trailing and Leading Spaces
anyClassDF = anyClassDF.applymap(lambda x: x.strip() if isinstance(x, str) else x)
anyClassFS = anyClassFS.applymap(lambda x: x.strip() if isinstance(x, str) else x)

#Create keys for finding
anyClassDF['key'] = anyClassDF['Folder'] + anyClassDF['Column']
anyClassFS['key'] = anyClassFS['Folder'] + anyClassFS['Column']
```

```python
print("number of columns in RPD for Any Class: " + str(len(anyClassDF)))
print("number of columns in FuncSpec for Any Class: " + str(len(anyClassFS)))
```

```python
resultAnyClassDF = anyClassFS[~anyClassFS.key.isin(anyClassDF.key)]
resultAnyClassDF.to_csv('anyClassMismatchList.csv', index = None, header=True)
```

```python
print("number of mismatch in funcSpec for Any Class: " + str(len(anyClassFS[~anyClassFS.key.isin(anyClassDF.key)])))
print("number of mistmatch in RPD for Any Class: " + str(len(anyClassDF[~anyClassDF.key.isin(anyClassFS.key)])))
```
